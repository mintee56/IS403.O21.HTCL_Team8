```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('./Cryptocurrencies/Initial Datasets/bitcoin_2019-01-01_2024-06-17.csv')
data2 = pd.read_csv('./Cryptocurrencies/Initial Datasets/dogecoin_2019-01-01_2024-06-17.csv')
data3 = pd.read_csv('./Cryptocurrencies/Initial Datasets/ethereum_2019-01-01_2024-06-17.csv')
```

```python
data1
```

```python
# Remove 'Start' column
data1 = data1.loc[:, ['End', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]
data2 = data2.loc[:, ['End', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]
data3 = data3.loc[:, ['End', 'Open', 'High', 'Low', 'Close', 'Volume', 'Market Cap']]

# Rename 'End' column to 'Date'
data1 = data1.rename(columns={'End': 'Date'})
data2 = data2.rename(columns={'End': 'Date'})
data3 = data3.rename(columns={'End': 'Date'})
```

```python
# Convert 'Date' column to datetime
data1['Date'] = pd.to_datetime(data1.Date)
data2['Date'] = pd.to_datetime(data2.Date)
data3['Date'] = pd.to_datetime(data3.Date)
```

```python
# Sort data ascending by Date
data1 = data1.sort_values(['Date'])
data2 = data2.sort_values(['Date'])
data3 = data3.sort_values(['Date'])

# Reset index
data1 = data1.reset_index(drop=True)
data2 = data2.reset_index(drop=True)
data3 = data3.reset_index(drop=True)
```

```python
# Export CSV
data1.to_csv('BTC.csv')
data2.to_csv('ETH.csv')
data3.to_csv('DOGE.csv')
```

```python
# Set 'Date' as index column
data1 = data1.set_index('Date')
data1.index = pd.to_datetime(data1.index)

data2 = data2.set_index('Date')
data2.index = pd.to_datetime(data2.index)

data3 = data3.set_index('Date')
data3.index = pd.to_datetime(data3.index)
```
