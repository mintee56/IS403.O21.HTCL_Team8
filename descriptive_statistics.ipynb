```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline

import seaborn as sns
```

```python
btc = pd.read_csv('./Cryptocurrencies/BTC.csv',
                  index_col='Date', parse_dates=True)
eth = pd.read_csv('./Cryptocurrencies/ETH.csv',
                  index_col='Date', parse_dates=True)
doge = pd.read_csv('./Cryptocurrencies/DOGE.csv',
                   index_col='Date', parse_dates=True)
```

```python
# BTC Box Plot
plt.figure(figsize=(6, 4))
sns.boxplot(x=btc['Close'], color='#e48644', width=0.5, linewidth=1)
plt.xlabel('BTC Close Price')
plt.show()
```

```python
# ETH Box Plot
plt.figure(figsize=(6, 4))
sns.boxplot(x=eth['Close'], color='#e48644', width=0.5, linewidth=1)
plt.xlabel('ETH Close Price')
plt.show()
```

```python
# DOGE Box Plot
plt.figure(figsize=(6, 4))
sns.boxplot(x=doge['Close'], color='#e48644', width=0.5, linewidth=1)
plt.xlabel('DOGE Close Price')
plt.show()
```

```python
# BTC Histogram
plt.figure(figsize=(6, 4))
sns.histplot(btc['Close'],color='#e48644', bins=20, kde=True)
plt.xlabel('BTC Close Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# ETH Histogram
plt.figure(figsize=(6, 4))
sns.histplot(eth['Close'],color='#e48644', bins=20, kde=True)
plt.xlabel('ETH Close Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# DOGE Histogram
plt.figure(figsize=(6, 4))
sns.histplot(doge['Close'],color='#e48644', bins=20, kde=True)
plt.xlabel('DOGE Close Price')
plt.ylabel('Frequency')
plt.show()
```
